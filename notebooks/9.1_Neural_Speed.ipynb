## Part 1: Action Potential Propagation

### The Speed Challenge
Neurons need to transmit signals quickly, but faster transmission requires more energy. Evolution has found two main solutions:

### Solution 1: Giant Axons (Invertebrates)
**Formula**: Speed ∝ √(diameter)

For unmyelinated axons:  
$$v = k \sqrt{d}$$

Where:
- v = conduction velocity (m/s)
- d = axon diameter (μm)
- k ≈ 5.5 (constant)

**Trade-off**: Speed requires HUGE axons (squid: 1mm diameter!)

### Solution 2: Myelination (Vertebrates)
**Formula**: Speed ∝ diameter (linear!)

For myelinated axons:  
$$v = 6d$$

Where:
- v = conduction velocity (m/s)
- d = axon diameter (μm)

**Advantage**: Same speed with 100× smaller diameter!

### Temperature Effects
**Formula**: Q₁₀ temperature coefficient

$$v_{T2} = v_{T1} \times Q_{10}^{(T2-T1)/10}$$

Where:
- Q₁₀ ≈ 2-3 for neural processes
- Doubling/tripling speed per 10°C

**Impact**: Cold reptiles have 3× slower reflexes!

```python
# Neural speed calculation functions
def calculate_speed_unmyelinated(diameter_um):
    """
    Calculate conduction velocity for unmyelinated axon
    Speed scales with square root of diameter
    
    Parameters:
    - diameter_um: Axon diameter in micrometers
    
    Returns:
    - velocity in m/s
    """
    k = 5.5  # Empirical constant
    velocity = k * np.sqrt(diameter_um)
    return velocity

def calculate_speed_myelinated(diameter_um):
    """
    Calculate conduction velocity for myelinated axon
    Speed scales linearly with diameter (saltatory conduction)
    
    Parameters:
    - diameter_um: Axon diameter in micrometers
    
    Returns:
    - velocity in m/s
    """
    velocity = 6 * diameter_um
    return velocity

def apply_temperature_effect(velocity_base, temp_current, temp_base=37, q10=2.5):
    """
    Apply temperature effect using Q10 coefficient
    
    Parameters:
    - velocity_base: Base velocity at temp_base
    - temp_current: Current temperature (°C)
    - temp_base: Base temperature (°C), default 37
    - q10: Temperature coefficient, default 2.5
    
    Returns:
    - velocity at current temperature
    """
    temp_diff = temp_current - temp_base
    velocity_adjusted = velocity_base * (q10 ** (temp_diff / 10))
    return velocity_adjusted

# Test examples
print("EXAMPLES:")
print("="*70)

# Giant squid axon
squid_diameter = 1000  # 1mm = 1000 μm!
squid_speed = calculate_speed_unmyelinated(squid_diameter)
print(f"\nGiant Squid Axon:")
print(f"  Diameter: {squid_diameter} μm (1 mm!)")
print(f"  Speed: {squid_speed:.1f} m/s")
print(f"  Type: Unmyelinated (needs huge size for speed)")

# Human motor neuron
human_diameter = 20  # 20 μm
human_speed = calculate_speed_myelinated(human_diameter)
print(f"\nHuman Motor Neuron:")
print(f"  Diameter: {human_diameter} μm (0.02 mm)")
print(f"  Speed: {human_speed:.1f} m/s")
print(f"  Type: Myelinated (50× smaller, similar speed!)")

# Temperature effect on lizard
lizard_speed_warm = 50  # m/s at 35°C
lizard_speed_cold = apply_temperature_effect(lizard_speed_warm, temp_current=15, temp_base=35)
print(f"\nLizard Neuron:")
print(f"  Speed at 35°C: {lizard_speed_warm:.1f} m/s (warm, active)")
print(f"  Speed at 15°C: {lizard_speed_cold:.1f} m/s (cold, sluggish)")
print(f"  Ratio: {lizard_speed_warm/lizard_speed_cold:.1f}× slower when cold!")

print("\n✓ Calculation functions ready!")
```

## Part 3: Interactive Speed Simulator

```python
def simulate_neural_transmission(species_name, temperature=None):
    """Simulate action potential transmission"""
    
    if species_name not in species_neural:
        print("Species not found")
        return
    
    sp = species_neural[species_name]
    
    # Use custom temperature if provided, else use body temp
    if temperature is None:
        temperature = sp['body_temp']
    
    # Calculate speed at given temperature
    base_speed = sp['speed_ms']
    actual_speed = apply_temperature_effect(base_speed, temperature, sp['body_temp'])
    
    # Calculate reflex time
    distance_m = sp['reflex_distance_cm'] / 100
    reflex_time_ms = (distance_m / actual_speed) * 1000
    
    # Create 4-panel visualization
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=('Conduction Velocity', 'Action Potential Animation',
                       'Myelinated vs Unmyelinated', 'Reflex Response Time'),
        specs=[[{'type': 'indicator'}, {'type': 'scatter'}],
               [{'type': 'bar'}, {'type': 'indicator'}]]
    )
    
    # 1. Speed gauge
    fig.add_trace(go.Indicator(
        mode="gauge+number",
        value=actual_speed,
        title={'text': f"{sp['type']}"},
        number={'suffix': " m/s"},
        gauge={
            'axis': {'range': [0, 150]},
            'bar': {'color': '#E74C3C' if sp['type'] == 'Myelinated' else '#3498DB'},
            'steps': [
                {'range': [0, 30], 'color': '#ECF0F1'},
                {'range': [30, 70], 'color': '#D5DBDB'},
                {'range': [70, 120], 'color': '#AED6F1'},
                {'range': [120, 150], 'color': '#85C1E9'}
            ]
        }
    ), row=1, col=1)
    
    # 2. Action potential propagation
    time_points = np.linspace(0, reflex_time_ms, 100)
    distance_points = (time_points / 1000) * actual_speed * 100  # in cm
    
    fig.add_trace(go.Scatter(
        x=time_points,
        y=distance_points,
        mode='lines',
        line=dict(color='#E74C3C' if sp['type'] == 'Myelinated' else '#3498DB', width=3),
        fill='tozeroy',
        name='Signal',
        showlegend=False
    ), row=1, col=2)
    
    # 3. Comparison
    comparison_diameter = sp['diameter_um']
    unmyel_speed = calculate_speed_unmyelinated(comparison_diameter)
    myel_speed = calculate_speed_myelinated(comparison_diameter)
    
    fig.add_trace(go.Bar(
        x=['Unmyelinated', 'Myelinated', f'{species_name}\n(Actual)'],
        y=[unmyel_speed, myel_speed, actual_speed],
        marker_color=['#3498DB', '#E74C3C', '#2ECC71'],
        text=[f"{unmyel_speed:.1f}", f"{myel_speed:.1f}", f"{actual_speed:.1f}"],
        textposition='outside',
        showlegend=False
    ), row=2, col=1)
    
    # 4. Reflex time
    fig.add_trace(go.Indicator(
        mode="number+delta",
        value=reflex_time_ms,
        title={'text': "Reflex Time"},
        number={'suffix': " ms"},
        delta={'reference': 10, 'relative': False}
    ), row=2, col=2)
    
    # Update layout
    fig.update_xaxes(title_text="Time (ms)", row=1, col=2)
    fig.update_yaxes(title_text="Distance (cm)", row=1, col=2)
    fig.update_yaxes(title_text="Speed (m/s)", row=2, col=1)
    
    fig.update_layout(
        height=700,
        title_text=f"<b>{species_name} Neural Transmission @ {temperature}°C</b>"
    )
    
    # Print summary
    print("\n" + "="*70)
    print(f"NEURAL ANALYSIS: {species_name}")
    print("="*70)
    print(f"Axon type: {sp['type']}")
    print(f"Diameter: {sp['diameter_um']} μm")
    print(f"Temperature: {temperature}°C (body temp: {sp['body_temp']}°C)")
    print(f"\nConduction velocity: {actual_speed:.1f} m/s")
    print(f"Reflex distance: {sp['reflex_distance_cm']} cm")
    print(f"Reflex time: {reflex_time_ms:.1f} ms")
    
    if sp['type'] == 'Myelinated':
        equiv_unmyel = (actual_speed / 5.5) ** 2
        print(f"\nTo achieve same speed unmyelinated would need:")
        print(f"  Diameter: {equiv_unmyel:.0f} μm ({equiv_unmyel/sp['diameter_um']:.0f}× larger!)")
    
    print(f"\n{sp['desc']}")
    print("="*70)
    
    fig.show()

# Interface
species_dropdown = Dropdown(
    options=sorted(species_neural.keys()),
    value='Human',
    description='Species:',
    style={'description_width': '80px'}
)

temp_slider = FloatSlider(
    value=37,
    min=5,
    max=40,
    step=1,
    description='Temp (°C):',
    style={'description_width': '80px'}
)

def update_sim(species, temp):
    simulate_neural_transmission(species, temp)

display(HTML("<h3>⚡ Neural Transmission Simulator</h3>"))
display(HTML("<p>Adjust temperature to see effects on transmission speed:</p>"))
interact(update_sim, species=species_dropdown, temp=temp_slider);
```

## Part 4: Myelination Advantage Calculator

```python
def compare_myelination(diameter_um):
    """Compare myelinated vs unmyelinated at same diameter"""
    
    unmyel_speed = calculate_speed_unmyelinated(diameter_um)
    myel_speed = calculate_speed_myelinated(diameter_um)
    
    # To match myelinated speed, what diameter unmyelinated needs?
    required_unmyel_diameter = (myel_speed / 5.5) ** 2
    
    # Cross-sectional areas
    area_myel = np.pi * (diameter_um/2)**2
    area_unmyel_equiv = np.pi * (required_unmyel_diameter/2)**2
    
    # Create visualization
    fig = make_subplots(
        rows=1, cols=3,
        subplot_titles=('Speed Comparison', 'Size Requirement', 'Space Efficiency'),
        specs=[[{'type': 'bar'}, {'type': 'bar'}, {'type': 'bar'}]]
    )
    
    # 1. Speed
    fig.add_trace(go.Bar(
        x=['Unmyelinated', 'Myelinated'],
        y=[unmyel_speed, myel_speed],
        marker_color=['#3498DB', '#E74C3C'],
        text=[f"{unmyel_speed:.1f} m/s", f"{myel_speed:.1f} m/s"],
        textposition='outside'
    ), row=1, col=1)
    
    # 2. Diameter comparison
    fig.add_trace(go.Bar(
        x=['Myelinated', 'Unmyelinated\n(same speed)'],
        y=[diameter_um, required_unmyel_diameter],
        marker_color=['#E74C3C', '#3498DB'],
        text=[f"{diameter_um:.0f} μm", f"{required_unmyel_diameter:.0f} μm"],
        textposition='outside'
    ), row=1, col=2)
    
    # 3. Space efficiency (how many fit?)
    space_ratio = area_unmyel_equiv / area_myel
    fig.add_trace(go.Bar(
        x=['Myelinated\naxons fit', 'Unmyelinated\naxons fit'],
        y=[space_ratio, 1],
        marker_color=['#E74C3C', '#3498DB'],
        text=[f"{space_ratio:.0f}×", "1×"],
        textposition='outside'
    ), row=1, col=3)
    
    fig.update_yaxes(title_text="Speed (m/s)", row=1, col=1)
    fig.update_yaxes(title_text="Diameter (μm)", row=1, col=2)
    fig.update_yaxes(title_text="Relative Count", row=1, col=3)
    
    fig.update_layout(
        height=500,
        title_text=f'<b>Myelination Advantage @ {diameter_um}μm Diameter</b>',
        showlegend=False
    )
    
    print("\n" + "="*70)
    print(f"MYELINATION ADVANTAGE ANALYSIS")
    print("="*70)
    print(f"Starting diameter: {diameter_um} μm\n")
    
    print(f"SPEEDS:")
    print(f"  Unmyelinated: {unmyel_speed:.1f} m/s")
    print(f"  Myelinated: {myel_speed:.1f} m/s")
    print(f"  Speed ratio: {myel_speed/unmyel_speed:.1f}× faster\n")
    
    print(f"SIZE EFFICIENCY:")
    print(f"  To match {myel_speed:.1f} m/s, unmyelinated needs:")
    print(f"    Diameter: {required_unmyel_diameter:.0f} μm ({required_unmyel_diameter/diameter_um:.0f}× larger)")
    print(f"    Area: {area_unmyel_equiv:.0f} μm² ({space_ratio:.0f}× more space)\n")
    
    print(f"PACKING ADVANTAGE:")
    print(f"  In same space, can fit:")
    print(f"    {space_ratio:.0f} myelinated axons OR")
    print(f"    1 unmyelinated axon (same speed)\n")
    
    print(f"RESULT: Myelination allows {space_ratio:.0f}× more neural pathways!")
    print("="*70)
    
    fig.show()

diameter_slider = IntSlider(
    value=20,
    min=1,
    max=50,
    step=1,
    description='Diameter (μm):',
    style={'description_width': '100px'}
)

display(HTML("<h3>🔬 Myelination Advantage</h3>"))
interact(compare_myelination, diameter_um=diameter_slider);
```

### Challenge 1: The Giant Squid Mystery 🦑

**Question**: Why does the giant squid have a 1mm diameter axon while humans use 20μm axons?

**Given**:
- Squid: 1000μm unmyelinated
- Human: 20μm myelinated
- Both achieve ~120 m/s

**Calculate**:
1. How many human axons fit in the space of one squid axon?
2. What's the energy cost difference?
3. Why didn't squid evolve myelination?

<details>
<summary>Solution</summary>

**Space analysis**:
- Squid area: π × (500μm)² = 785,000 μm²
- Human area: π × (10μm)² = 314 μm²
- Ratio: 2,500 human axons fit in one squid axon!

**Energy cost**:
- Maintaining ion gradients ∝ surface area
- Squid: 2× diameter → 4× energy
- But: 2,500 separate axons also costly!

**Why no myelination in invertebrates?**
1. Myelination requires glial cells (complex!)
2. Evolved in vertebrates only
3. Squid solution works: one fast fiber for escape
4. Don't need thousands of connections like vertebrates

**Lesson**: Different solutions to same problem based on evolutionary history!
</details>

```python
# Challenge 1 workspace
print("CHALLENGE 1: GIANT SQUID vs HUMAN")
print("="*70)

squid = species_neural['Squid (Giant Axon)']
human = species_neural['Human']

print(f"SQUID GIANT AXON:")
print(f"  Diameter: {squid['diameter_um']} μm")
print(f"  Type: {squid['type']}")
print(f"  Speed: {squid['speed_ms']:.1f} m/s\n")

print(f"HUMAN MOTOR NEURON:")
print(f"  Diameter: {human['diameter_um']} μm")
print(f"  Type: {human['type']}")
print(f"  Speed: {human['speed_ms']:.1f} m/s\n")

# Calculate how many fit
squid_area = np.pi * (squid['diameter_um']/2)**2
human_area = np.pi * (human['diameter_um']/2)**2
ratio = squid_area / human_area

print(f"SPACE EFFICIENCY:")
print(f"  Squid axon area: {squid_area:,.0f} μm²")
print(f"  Human axon area: {human_area:,.0f} μm²")
print(f"  Ratio: {ratio:,.0f} human axons fit in one squid axon!\n")

print(f"EVOLUTIONARY TRADE-OFF:")
print(f"  Squid: Simple, giant fiber for rapid escape")
print(f"  Human: Complex network with {ratio:,.0f}× more connections")
print(f"  Both achieve ~{squid['speed_ms']:.0f} m/s, different strategies!")

print("\n" + "="*70)
print("YOUR ANALYSIS:")
print("-"*70)
print("Which strategy is 'better'? Depends on what you need!")
print("="*70)
```

### Challenge 2: Cold-Blooded Reflexes 🦎

**Scenario**: A lizard at 15°C vs 35°C

**Given**:
- Base speed at 35°C: 60 m/s
- Q₁₀ = 2.5
- Reflex distance: 25 cm

**Calculate**:
1. Speed at 15°C?
2. Reflex time at both temperatures?
3. Can it escape a predator?

<details>
<summary>Solution</summary>

**Speed at 15°C**:
- Temp drop: 20°C (two 10°C intervals)
- Speed = 60 × (2.5)^(-20/10) = 60 × (1/6.25) = 9.6 m/s

**Reflex times**:
- At 35°C: (0.25m / 60m/s) × 1000 = 4.2 ms
- At 15°C: (0.25m / 9.6m/s) × 1000 = 26 ms

**Predator escape**:
- Warm lizard: Fast enough (~4ms response)
- Cold lizard: 6× slower! Vulnerable!

**This is why lizards bask in the sun!**
</details>

```python
# Challenge 2 workspace
print("CHALLENGE 2: TEMPERATURE EFFECTS ON LIZARD")
print("="*70)

base_speed = 60  # m/s at 35°C
reflex_distance = 0.25  # 25cm in meters
q10 = 2.5

temps = [35, 30, 25, 20, 15]

print(f"Base conditions: {base_speed} m/s at 35°C\n")
print(f"{'Temp (°C)':<12}{'Speed (m/s)':<15}{'Reflex (ms)':<15}{'Status':<20}")
print("-"*70)

for temp in temps:
    speed = apply_temperature_effect(base_speed, temp, 35, q10)
    reflex_ms = (reflex_distance / speed) * 1000
    
    if temp >= 30:
        status = "✓ Active, fast"
    elif temp >= 25:
        status = "⚠ Sluggish"
    elif temp >= 20:
        status = "⚠ Very slow"
    else:
        status = "✗ Vulnerable!"
    
    print(f"{temp:<12}{speed:<15.1f}{reflex_ms:<15.1f}{status:<20}")

print("-"*70)
print(f"\nSpeed range: {apply_temperature_effect(base_speed, 15, 35, q10):.1f} to {base_speed:.1f} m/s")
print(f"That's a {base_speed/apply_temperature_effect(base_speed, 15, 35, q10):.1f}× difference!")
print("\nThis is why reptiles bask in the sun before hunting!")
print("="*70)
```
